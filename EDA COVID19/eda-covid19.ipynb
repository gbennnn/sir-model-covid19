```python
# Impor Library yang Dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Untuk menampilkan semua kolom sampai 200
pd.set_option('display.max_columns', 200) 

# Mengatur gaya plot untuk visualisasi yang konsisten
plt.style.use('ggplot')
```

```python
# Memuat dataset
file_path = "../dataset/covid19-indonesia.csv"
df = pd.read_csv(file_path)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Filter the dataframe for the specified date range
start_date = '2021-01-01'
end_date = '2022-01-31'
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
df_filtered = df.loc[mask]

# Calculate the average number of new cases
average_new_cases = df_filtered['New Cases'].mean()
print(f"Average number of new cases from January 2021 to January 2022: {average_new_cases}")
```

```python
# Calculate the average number of recovered cases.
average_recovered_cases = df_filtered['New Recovered'].mean()
print(f"Average number of recovered cases from January 2021 to January 2022: {average_recovered_cases}")
```

```python
# Calculate the average number of deaths per day
average_daily_deaths = df_filtered['New Deaths'].mean()
print(f"Average number of deaths per day from January 2021 to January 2022: {average_daily_deaths}")
```

```python
# Calculate the percentage of recovered individuals per day
df_filtered['Recovery Rate'] = (df_filtered['New Recovered'] / df_filtered['New Cases']) * 100

# Display the dataframe with the new 'Recovery Rate' column
df_filtered[['Date', 'New Cases', 'New Recovered', 'Recovery Rate']]
```

```python
# Remove rows with inf or NaN values in 'Recovery Rate'
df_filtered_cleaned = df_filtered.replace([np.inf, -np.inf], np.nan).dropna(subset=['Recovery Rate'])

# Calculate the average recovery rate
average_recovery_rate = df_filtered_cleaned['Recovery Rate'].mean()
print(f"Average recovery rate per day from January 2021 to January 2022: {average_recovery_rate:.2f}%")
```
